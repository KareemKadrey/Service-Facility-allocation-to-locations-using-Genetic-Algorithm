```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math


# Problem parameters
num_vaccine_centers = 5 ## The number of available vaccine centers.
num_populations = 10    ## The number of populations that need to be assigned to vaccine centers.
max_capacity = [random.randint(120, 200) for _ in range(num_vaccine_centers)]  ## The maximum capacity of each vaccine center, i.e., the maximum number of populations that a center can accommodate.
min_capacity = random.randint(20, 100)
max_distance = 10       ## The maximum distance between a population and its assigned vaccine center.
max_num_centers = 1    ## The maximum number of vaccine centers that can be assigned to a single population.
population_locations = np.random.rand(num_populations, 2)  # Random population locations ## A 2D numpy array representing the randomly generated locations of the populations.
vaccine_center_locations = np.random.rand(num_vaccine_centers, 2)  # Random vaccine center locations ## A 2D numpy array representing the randomly generated locations of the vaccine centers.
max_population_distance = 5  # Maximum distance between a population and its assigned vaccine center ## The maximum allowed distance between a population and its assigned vaccine center.
min_population_distance = 2  # Minimum distance between a population and its assigned vaccine center
max_center_distance = 10
max_num_populations = 2  # Maximum number of populations that can be assigned to a single vaccine center

# Genetic Algorithm parameters
population_size = 50   ## The size of the population in the Genetic Algorithm, i.e., the number of solutions in each generation.
num_generations = 100  ## The number of generations in the Genetic Algorithm, i.e., the number of iterations the algorithm will run.
elitism_rate = 0.1



# Objective function
def objective_function(solution):
    # Calculate the objective value of a solution
    total_distance = 0
    for j in range(num_populations):
        assigned_center = np.argmax(solution[j])
        total_distance += calculate_distance(population_locations[j], vaccine_center_locations[assigned_center])

    return total_distance

# Modified fitness function incorporating the objective function
def fitness_function(solution):
    # Evaluate the fitness value of a solution using the objective function
    objective_value = objective_function(solution)
    return objective_value

def calculate_distance(loc1, loc2):
    # Calculate Euclidean distance between two locations
    return np.sqrt(np.sum((loc1 - loc2)**2))


def initialize_population():
    # Initialize a random population of solutions
    population = []
    for _ in range(population_size):
        solution = np.zeros((num_populations, num_vaccine_centers))
        for j in range(num_populations):
            assigned_centers = random.sample(range(num_vaccine_centers), random.randint(1, max_num_centers))
            for center in assigned_centers:
                solution[j][center] = 1
        population.append(solution)
    return population

def crossover(parent1, parent2):
    # Perform crossover between two parent solutions to generate a child solution
    child = np.zeros((num_populations, num_vaccine_centers))
    for j in range(num_populations):
        if random.random() < 0.5:
            child[j] = parent1[j]
        else:
            child[j] = parent2[j]
    return child

def mutate(solution, generation):
    # Calculate the mutation rate based on the current generation
    max_mutation_rate = 0.1  # Maximum mutation rate
    min_mutation_rate = 0.01  # Minimum mutation rate
    total_generations = num_generations  # Total number of generations
    mutation_rate = max_mutation_rate - (max_mutation_rate - min_mutation_rate) * (generation / total_generations)

    for j in range(num_populations):
        if random.random() < mutation_rate:
            # Perform mutation on a solution by randomly changing an assignment
            assigned_centers = np.where(solution[j] == 1)[0]
            if assigned_centers.size > 0:
                center_to_remove = random.choice(assigned_centers)
                solution[j][center_to_remove] = 0
            assigned_center = random.randint(0, num_vaccine_centers-1)
            solution[j][assigned_center] = 1


    # Apply first constraint: Maximum distance between a population and its assigned vaccine center
    for j in range(num_populations):
        assigned_center = np.argmax(solution[j])
        distance = calculate_distance(population_locations[j], vaccine_center_locations[assigned_center])
        if distance > max_population_distance:    ## max_population_distance = 5
            # Find a new center within the distance constraint
            centers_within_distance = np.where(
                np.array([calculate_distance(population_locations[j], loc) for loc in vaccine_center_locations])
                <= max_population_distance)[0]
            if centers_within_distance.size > 0:
                new_center = random.choice(centers_within_distance)
                solution[j] = np.zeros(num_vaccine_centers)
                solution[j][new_center] = 1
                

    return solution

def rank_based_selection(population, fitness_values, selection_probabilities, population_size):
    ranked_indices = np.argsort(fitness_values)
    ranked_population = [population[index] for index in ranked_indices]
    
    parents = []
    for _ in range(population_size):
        parent_index = np.random.choice(range(len(population)), p=selection_probabilities)
        parents.append(ranked_population[parent_index])
    
    return parents

def tournament_selection(population, fitness_values, tournament_size):
    selected_parents = []
    for _ in range(len(population)):
        tournament_indices = random.sample(range(len(population)), tournament_size)
        tournament_fitness = [fitness_values[i] for i in tournament_indices]
        winner_index = np.argmin(tournament_fitness)
        selected_parents.append(population[tournament_indices[winner_index]])
    return selected_parents


def roulette_wheel_selection(population, fitness_values, population_size):
    selection_probabilities = [1 / f for f in fitness_values]
    selection_probabilities = np.array(selection_probabilities) / sum(selection_probabilities)
    selected_parents_indices = np.random.choice(range(len(population)), size=population_size, p=selection_probabilities)
    selected_parents = [population[i] for i in selected_parents_indices]
    return selected_parents


def local_search(solution):
    # Perform local search to optimize the solution further
    best_solution = solution.copy()
    best_fitness = fitness_function(best_solution)
    improvement = True

    while improvement:
        improvement = False
        for i in range(num_populations - 1):
            for j in range(i + 1, num_populations):
                new_solution = best_solution.copy()
                # Swap the assignment of populations i and j to their vaccine centers
                new_solution[i], new_solution[j] = new_solution[j], new_solution[i]
                new_fitness = fitness_function(new_solution)
                if new_fitness < best_fitness:
                    best_solution = new_solution.copy()
                    best_fitness = new_fitness
                    improvement = True

    return best_solution



def enforce_constraints(solution):
    # Enforce constraints on the solution
    ## apply second constraint
    for j in range(num_populations):
        assigned_centers = np.where(solution[j] == 1)[0]
        if assigned_centers.size > max_num_centers:    ## max_num_centers = 3
            centers_to_remove = random.sample(list(assigned_centers), assigned_centers.size - max_num_centers)
            for center in centers_to_remove:
                solution[j][center] = 0
                
    ## apply third constraint
    capacities = np.sum(solution, axis=0)
    for center in range(num_vaccine_centers):
        if capacities[center] > max_capacity[center]:
            populations = np.where(solution[:, center] == 1)[0]
            populations_to_remove = random.sample(list(populations), capacities[center] - max_capacity)
            for population in populations_to_remove:
                solution[population][center] = 0
                          
    

    return solution


## apply fourth constraint
def balance_assignment(solution):
    # Balance the assignment of populations to vaccine centers
    capacities = np.sum(solution, axis=0)
    min_capacity = int(np.min(capacities))
    max_capacity = int(np.max(capacities))
    centers_below_min = np.where(capacities < min_capacity)[0]
    centers_above_max = np.where(capacities > max_capacity)[0]

    for center in centers_below_min:
        populations = np.where(solution[:, center] == 1)[0]
        unassigned_populations = np.where(np.sum(solution, axis=1) == 0)[0]
        if unassigned_populations.size > 0:
            population_to_assign = random.choice(unassigned_populations)
            solution[population_to_assign][center] = 1

    for center in centers_above_max:
        populations = np.where(solution[:, center] == 1)[0]
        populations_to_remove = random.sample(list(populations), capacities[center] - max_capacity)
        for population in populations_to_remove:
            solution[population][center] = 0

    return solution

## apply fifth constraint 
def population_accessibility(solution):
    # Enforce population accessibility constraint
    for j in range(num_populations):
        assigned_centers = np.where(solution[j] == 1)[0]
        if assigned_centers.size == 0:
            accessible_centers = np.where(
                np.array([calculate_distance(population_locations[j], loc) for loc in vaccine_center_locations])
                <= max_distance)[0]
            if accessible_centers.size > 0:
                new_center = random.choice(accessible_centers)
                solution[j] = np.zeros(num_vaccine_centers)
                solution[j][new_center] = 1

    return solution


## apply six constraint
def center_capacity_constraint(solution):
    # Enforce minimum and maximum capacity constraints for vaccine centers
    capacities = np.sum(solution, axis=0)
    min_capacity = int(np.min(capacities))
    max_capacity = int(np.max(capacities))
    for center in range(num_vaccine_centers):
        if capacities[center] < min_capacity:
            populations = np.where(solution[:, center] == 0)[0]
            populations_to_assign = random.sample(list(populations), min_capacity - capacities[center])
            for population in populations_to_assign:
                solution[population][center] = 1
        elif capacities[center] > max_capacity:
            populations = np.where(solution[:, center] == 1)[0]
            populations_to_remove = random.sample(list(populations), capacities[center] - max_capacity)
            for population in populations_to_remove:
                solution[population][center] = 0

    return solution

# Constraint 7: Maximum distance between vaccine centers
def max_center_dist(solution):
    for center1 in range(num_vaccine_centers):
        for center2 in range(center1 + 1, num_vaccine_centers):
            distance = calculate_distance(vaccine_center_locations[center1], vaccine_center_locations[center2])
            if distance > max_center_distance:
                # Find a new location for one of the centers within the distance constraint
                new_location = None
                centers_within_distance = np.where(
                    np.array([calculate_distance(loc, vaccine_center_locations[center1]) for loc in vaccine_center_locations])
                    <= max_center_distance)[0]
                if centers_within_distance.size > 0:
                    new_location = random.choice(centers_within_distance)
                centers_within_distance = np.where(
                    np.array([calculate_distance(loc, vaccine_center_locations[center2]) for loc in vaccine_center_locations])
                    <= max_center_distance)[0]
                if centers_within_distance.size > 0:
                    new_location = random.choice(centers_within_distance)
                if new_location is not None:
                    # Update the location of one of the centers
                    solution[:, center2] = solution[:, center1]
                    vaccine_center_locations[center2] = vaccine_center_locations[center1]

    return solution

def display_statistics(solution):
    num_populations_assigned = np.sum(solution, axis=0)
    total_capacity = np.sum(num_populations_assigned)
    num_vaccine_centers = len(num_populations_assigned)
    average_assigned = np.mean(num_populations_assigned)
    min_assigned = np.min(num_populations_assigned)
    max_assigned = np.max(num_populations_assigned)


    print("Number of populations assigned to each vaccine center:")
    print('\n'.join([f"Vaccine Center {i + 1}: {populations}" for i, populations in enumerate(num_populations_assigned)]))

    print()
    print("Total capacity of each vaccine center:")
    print('\n'.join([f"Vaccine Center {i + 1}: {populations}/{max_capacity[i]} (assigned/maximum)" for i, populations in enumerate(num_populations_assigned)]))

    print()
    print("Minimum number of populations assigned to a vaccine center:", min_assigned)
    print("Maximum number of populations assigned to a vaccine center:", max_assigned)
    print()
    print("Total capacity of all vaccine centers:", total_capacity)
    print("Percentage of maximum capacity utilized by each vaccine center:")
    print('\n'.join([f"Vaccine Center {i + 1}: {(populations / max_capacity[i]) * 100:.2f}%" for i, populations in enumerate(num_populations_assigned)]))
    print()
    print("Average number of populations assigned per vaccine center:", average_assigned)
    print()
    print("Distribution of populations across vaccine centers:")
    for center, populations in enumerate(num_populations_assigned):
        percentage_assigned = (populations / total_capacity) * 100
        print(f"Vaccine Center {center + 1}: {percentage_assigned:.2f}%")
    print('-' * 100)

def genetic_algorithm(selection_method):
    # Main Genetic Algorithm function
    population = initialize_population()
    best_fitness_values = []
    best_individuals = []
    

    for generation in range(num_generations):
        # Calculate fitness values for the current population
        fitness_values = [fitness_function(solution) for solution in population]
        best_fitness = min(fitness_values)
        best_fitness_values.append(best_fitness)
        best_solution_index = np.argmin(fitness_values)
        best_individuals.append(population[best_solution_index])

        # Rank individuals based on fitness
        num_individuals = len(population)
        selection_probabilities = [(2 * (num_individuals - rank)) / (num_individuals * (num_individuals + 1)) for rank in range(num_individuals)]
        selection_probabilities = np.array(selection_probabilities) / sum(selection_probabilities)  # Normalize the selection probabilities
        
        # Perform selection using the specified method
        if selection_method == 'tournament':
            parents = tournament_selection(population, fitness_values, tournament_size=5)
        elif selection_method == 'roulette_wheel':
            parents = roulette_wheel_selection(population, fitness_values, population_size)
        elif selection_method == 'rank_based':
            parents = rank_based_selection(population, fitness_values, selection_probabilities, population_size)
        else:
            raise ValueError(f"Invalid selection method: {selection_method}")



        # Perform crossover and mutation to create the next generation
        next_generation = []
        for i in range(0, population_size, 2):
            parent1 = parents[i]
            parent2 = parents[i+1]
            child1 = crossover(parent1, parent2)
            child2 = crossover(parent2, parent1)
            child1 = mutate(child1, generation)  # Pass the current generation to the mutate function
            child2 = mutate(child2, generation)  # Pass the current generation to the mutate function
            child1 = enforce_constraints(child1)
            child2 = enforce_constraints(child2)
            child1 = balance_assignment(child1)
            child2 = balance_assignment(child2)
            child1 = population_accessibility(child1)
            child2 = population_accessibility(child2)
            child1 = center_capacity_constraint(child1)
            child2 = center_capacity_constraint(child2)
            child1 = max_center_dist(child1)
            child2 = max_center_dist(child2)
            next_generation.extend([child1, child2])
            
        # Introduce elitism by preserving the best individuals from the current generation
        num_elites = int(population_size * elitism_rate)
        elites_indices = np.argsort(fitness_values)[:num_elites]
        elites = [population[index] for index in elites_indices]
        next_generation[:num_elites] = elites

        population = next_generation

    # Calculate fitness values for the final population
    fitness_values = [fitness_function(solution) for solution in population]

    # Select the best solution as the final result
    best_solution_index = np.argmin(fitness_values)
    best_solution = population[best_solution_index]
    best_fitness = fitness_values[best_solution_index]
    
    # Perform local search to optimize the best solution
    best_solution = local_search(best_solution)
    display_statistics(best_solution)

    # Print the best individual and its chromosome values in each generation
    for generation in range(num_generations):
        print("Generation", generation+1)
        print()
        best_fitness = best_fitness_values[generation]
        print("Fitness:", best_fitness)
        print()
        best_individual = best_individuals[generation]
        print("Best Individual:")
        print(best_individual)
        assigned_centers = np.argmax(best_individual, axis=1)
        for j in range(num_populations):
            if j > 0:
                print(",", end=" ")
            if best_individual[j][assigned_centers[j]] == 1:
                print("Population", j+1, "assigned to Vaccine Center", assigned_centers[j]+1, "(Open)", end="")
            else:
                print("Population", j+1, "assigned to Vaccine Center", assigned_centers[j]+1, "(Closed)", end="")
        print('-'*100)
        print()

    # Print the value of the objective function for the final solution
    print("Final Objective Function Value:", best_fitness)

    return best_solution, best_fitness_values, best_individuals  # Modify the return statement

# Visualization
def visualize_solution(population_locations, vaccine_center_locations, best_solution):
    assigned_centers = np.argmax(best_solution, axis=1)

    plt.figure(figsize=(10, 8))
    plt.scatter(population_locations[:, 0], population_locations[:, 1], c=assigned_centers, cmap='tab20', label='Population')
    plt.scatter(vaccine_center_locations[:, 0], vaccine_center_locations[:, 1], marker='s', color='red', label='Vaccine Center')
    plt.title('Assignment of Populations to Vaccine Centers')
    plt.xlabel('Population X-coordinate')
    plt.ylabel('Population Y-coordinate')
    plt.legend()
    plt.show()

# Performance metrics
def calculate_performance_metrics(best_solution):
    assigned_centers = np.argmax(best_solution, axis=1)
    total_distance = objective_function(best_solution)
    total_capacity = np.sum(best_solution)
    num_open_centers = np.sum(np.sum(best_solution, axis=0) > 0)

    average_distance = total_distance / num_populations
    utilization_rate = total_capacity / (num_populations * max_capacity)
    return average_distance, utilization_rate, num_open_centers

    
def plot_convergence_curves(best_fitness_values, average_distances, utilization_rates):
    generations = range(1, num_generations + 1)

    # Plot best fitness value convergence curve
    plt.figure(figsize=(10, 4))
    plt.plot(generations, best_fitness_values)
    plt.title('Best Fitness Value Convergence Curve')
    plt.xlabel('Generation')
    plt.ylabel('Best Fitness Value')
    plt.show()

    # Plot average distance convergence curve
    plt.figure(figsize=(10, 4))
    plt.plot(generations, average_distances)
    plt.title('Average Distance Convergence Curve')
    plt.xlabel('Generation')
    plt.ylabel('Average Distance')
    plt.show()

    # Plot utilization rate convergence curve
    plt.figure(figsize=(10, 4))
    plt.plot(generations, utilization_rates)
    plt.title('Utilization Rate Convergence Curve')
    plt.xlabel('Generation')
    plt.ylabel('Utilization Rate')
    plt.show()
    
def visualize_assigned_populations(best_solution):
    assigned_centers = np.argmax(best_solution, axis=1)
    population_counts = np.sum(best_solution, axis=0)

    plt.bar(range(1, num_vaccine_centers+1), population_counts)
    plt.title('Distribution of Assigned Populations among Vaccine Centers')
    plt.xlabel('Vaccine Center')
    plt.ylabel('Population Count')
    plt.show()

# Perform GA with different selection methods
selection_methods = ['tournament', 'roulette_wheel', 'rank_based']
fitness_values_per_selection = []

for selection_method in selection_methods:
    print(f"Now we will apply {selection_method} selection method")
    best_solution, best_fitness_values, best_individuals = genetic_algorithm(selection_method)
    fitness_values_per_selection.append(best_fitness_values)

# Plot the fitness values for each selection method
plt.figure(figsize=(10, 8))
for i, selection_method in enumerate(selection_methods):
    plt.plot(range(1, num_generations + 1), fitness_values_per_selection[i], label=selection_method)
plt.xlabel('Generation')
plt.ylabel('Fitness Value')
plt.title('Comparison of Selection Methods')
plt.legend()
plt.show()

# Visualize the best solution
visualize_solution(population_locations, vaccine_center_locations, best_solution)
visualize_assigned_populations(best_solution)


# Calculate and display performance metrics
average_distance, utilization_rate, num_open_centers = calculate_performance_metrics(best_solution)
print("Average Distance:", average_distance)
print("Utilization Rate:", utilization_rate)
print("Number of Open Centers:", num_open_centers)

# Calculate additional metrics for convergence curves
average_distances = []
utilization_rates = []
for generation in range(num_generations):
    solution = best_individuals[generation]
    avg_distance, util_rate, _ = calculate_performance_metrics(solution)
    average_distances.append(avg_distance)
    utilization_rates.append(util_rate)

# Plot convergence curves
plot_convergence_curves(best_fitness_values, average_distances, utilization_rates)


# Print the best solution
print("Best Solution:")
assigned_centers = np.argmax(best_solution, axis=1)
for j in range(num_populations):
    if best_solution[j][assigned_centers[j]] == 1:
        print("Population", j+1, "assigned to Vaccine Center", assigned_centers[j]+1, "(Open)")
    else:
        print("Population", j+1, "assigned to Vaccine Center", assigned_centers[j]+1, "(Closed)")
for center in range(num_vaccine_centers):
    if np.sum(best_solution[:, center]) > 0:
        print("Vaccine Center", center+1, ": will open")
    else:
        print("Vaccine Center", center+1, ": will not open")
```
